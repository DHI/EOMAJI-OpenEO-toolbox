```python
import openeo
```

```python
connection = openeo.connect("https://openeo.cloud")
connection.authenticate_oidc()
```

```python
# Load high-res and low-res data from OpenEO Platform
aoi = {
    "type": "Polygon",
    "coordinates": [[
      [
        8.1733155,
        56.554395
      ],
      [
        8.1733155,
        56.554395
      ],
      [
        8.1671337,
        56.4854867
      ],
      [
        8.2378811,
        56.5051875
      ],
      [
        8.2323862,
        56.5464504
      ],
      [
        8.1733155,
        56.554395
      ]
    ]]
}
start_date = "2024-09-17"
end_date = "2024-09-19"

sentinel2_l2a_cube = connection.load_collection(
    "SENTINEL2_L2A",
    spatial_extent=aoi,
    temporal_extent=[start_date, end_date],
    #bands=["B04", "B03", "B02"], 
)
sentinel3_lst = connection.load_collection(
    "SENTINEL3_SLSTR",
    spatial_extent=aoi,
   temporal_extent=[start_date, end_date],
   bands=['S8'],  
)
```

```python
udf = openeo.UDF.from_file("../openeo_workflows/dms/udf_datacube_context.py")
```

```python
results = sentinel2_l2a_cube.apply(process=udf, context = {"low_res_datacube":sentinel3_lst})
```

```python
results.download("output.tiff")
```
